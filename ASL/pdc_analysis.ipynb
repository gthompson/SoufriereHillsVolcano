```python
import os
from glob import glob
import obspy
seisandir='/raid/data/Montserrat/seisan/WAV/MVOE_/2003/06/'
seisanfiles=sorted(glob(os.path.join(seisandir,'*MVO*')))
for seisanfile in seisanfiles[0:10]:
    print(seisanfile)
    st=obspy.read(seisanfile)
    mblgz=st.select(location='Z').select(station='MBRY')[0]
    mblgz_low = mblgz.copy().filter('bandpass', freqmin=0.05, freqmax=0.5)
    mblgz_low.stats.location='ZL'
    mblgz_high = mblgz.copy().filter('highpass', freq=3.0)
    mblgz_high.stats.location='ZH'
    st2 = obspy.Stream().append(mblgz)
    st2.append(mblgz_low)
    st2.append(mblgz_high)
    st2.plot(equal_scale=False);
    
```
