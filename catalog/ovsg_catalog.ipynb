```python
# Download OVSG catalog around Montserrat
import os
import urllib.request
URL = 'http://services.seismes-antilles.fr/fdsnws/event/1/query?start=1990-01-01T00:00:00&end=2019-12-31T23:59:59&mindepth=-30&maxdepth=30&orderby=time&format=xml&lat=16.7425&lon=-62.1874&minradius=0.0&maxradius=0.2'
HOME = os.getenv('HOME')
QuakeMLfile = '%s/DATA/MVO/montserrat_catalog_ovsg.xml' % HOME
with urllib.request.urlopen(URL) as f:
    f2 = open(QuakeMLfile, 'w')
    s = f.read().decode('utf-8')
    f2.write(s)
    f2.close()
    print(os.getcwd())

# read downloaded QuakeML file
from obspy import read_events
cat = read_events(QuakeMLfile)
"""
cat = cat.filter("time > 1992-01-01T00:00", \
                  "time < 2010-01-11T23:59", \
                  "latitude > 16.7", \
                  "latitude < 16.8", \
                  "longitude < -62.14", \
                  "longitude > -62.23")
"""
# Montserrat: 16.7425Â° N, 62.1874
cat.write(QuakeMLfile.replace('.xml','.kml'),'KML')

print(cat.__str__(print_all=True))
```
